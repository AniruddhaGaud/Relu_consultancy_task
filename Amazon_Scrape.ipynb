```python
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
# take url from sheet...

sheet_url = 'https://docs.google.com/spreadsheets/d/1BZSPhk1LDrx8ytywMHWVpCqbm8URTxTJrIRkD7PnGTM/edit#gid=0'

# Editing url according to condition......

url_1 = sheet_url.replace("/edit#gid=", "/export?format=csv&gid=")
df = pd.read_csv(url_1)
df = df.reset_index()
driver = webdriver.Chrome(ChromeDriverManager().install())
invalid_urls, valid_urls, FINAL_DATA = [], [], []
payload = {}
count = 0
for index, row in df.iterrows():
    if count == 0:
        start = time.time() # importing time... 
    if count == 100:
        end = time.time()
        print("Time Taken for Each 100 URLS : ",end-start)
        count = 0
    count += 1
    asin, country = row['Asin'], row['country']
    amazon_url = f'https://www.amazon.{country}/dp/{asin}'
    driver.get(amazon_url)
    try: 
        title = driver.find_element_by_xpath('//*[@id="productTitle"]') # Extracting data from given url....
        if title:
            valid_urls.append(amazon_url)
            titlee = driver.find_element_by_xpath('//*[@id="productTitle"]')
            pricee = driver.find_element_by_xpath('//*[@id="a-autoid-3-announce"]')
            detail = driver.find_element_by_xpath('//div[@id="detailBulletsWrapper_feature_div"]')
            img_url = driver.find_element_by_xpath("//div[@id='img-canvas']/img").get_attribute("src")
            print("URL FOUND : ",amazon_url)
            payload['Price'] = pricee.text.split('\n')[-1]
            payload['Title'] = titlee.text
            payload['Image_URL'] = img_url
            payload['DETAILS'] = detail.text
            FINAL_DATA.append(payload)  # storing every extracted data in payload....
            print(payload)
            print("\n****************\n")
    except NoSuchElementException:
        invalid_urls.append(amazon_url)
        print(f"\n\n****URL NOT FOUND {amazon_url}*****\n\n")
driver.close()
print(FINAL_DATA)
print("invalid_urls : ",invalid_urls)
print("valid urls : ",valid_urls)
```
